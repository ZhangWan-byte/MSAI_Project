# copy to sequence

```python
with open("./data/cdr_copy_20221128.txt", "rb") as f:
    data = f.readlines()

f.close()
```

```python
res = []
pdb = []
i = 0

t1 = time.time()
while i<=(len(data)-1):
    if i%100==0:
        print("i: ", i)
    
    if len(data[i].decode().split())!=2 or (data[i].decode().split()[0]!="Summary" and data[i].decode().split()[1]!="file"):
        pdb.append(data[i].decode())
    else:
        res.append(pdb)
        pdb = []
    
    i += 1
t2 = time.time()

t2-t1
```

```python
res_1 = {}
entry = {}
pdb_name = None

for i in tqdm(range(len(res))):
    for j in range(len(res[i])):
        line = res[i][j].split()
        if len(line[0])==4 and line[0][0].isdigit():
            pdb_name = line[0]
            HeavyLightChains = line[-1].split(":")[0]#.upper()
#             if len(HeavyLightChains)==2:
#                 if HeavyLightChains[0].upper() == HeavyLightChains[1].upper():
#                     HeavyLightChains[0] = HeavyLightChains[0].upper()
#                     HeavyLightChains[1] = HeavyLightChains[1].lower()
            entry[pdb_name] = {HeavyLightChains:{}}
        elif len(line[0])==3 and line[0][0]=="F":
            HeavyLightChains = line[-1].split(":")[0]#.upper()
#             if len(HeavyLightChains)==2:
#                 if HeavyLightChains[0].upper() == HeavyLightChains[1].upper():
#                     HeavyLightChains[0] = HeavyLightChains[0].upper()
#                     HeavyLightChains[1] = HeavyLightChains[1].lower()
            entry[pdb_name][HeavyLightChains] = {}
        elif line[0]=="L1:":
            entry[pdb_name][HeavyLightChains]["L1"]=line[1] if len(line)==2 else None
        elif line[0]=="L2:":
            entry[pdb_name][HeavyLightChains]["L2"]=line[1] if len(line)==2 else None
        elif line[0]=="L3:":
            entry[pdb_name][HeavyLightChains]["L3"]=line[1] if len(line)==2 else None
        elif line[0]=="H1:":
            entry[pdb_name][HeavyLightChains]["H1"]=line[1] if len(line)==2 else None
        elif line[0]=="H2:":
            entry[pdb_name][HeavyLightChains]["H2"]=line[1] if len(line)==2 else None
        elif line[0]=="H3:":
            entry[pdb_name][HeavyLightChains]["H3"]=line[1] if len(line)==2 else None
```
